```python
def convert_price(price):
    # Verificar si el precio es un número (int o float)
    if isinstance(price, (int, float)):
        return price
    
    # Eliminar el símbolo de euro si es una cadena
    if isinstance(price, str):
        price = price.replace('€', '')
        
        # Manejar las unidades y convertir a número
        if 'bn' in price:
            price = float(price.replace('bn', '')) * 1e9
        elif 'm' in price:
            price = float(price.replace('m', '')) * 1e6
        elif 'k' in price:
            price = float(price.replace('k', '')) * 1e3
        else:
            price = float(price)  # Para casos como '0'
        
        return int(price)
    
    # Si el valor no es un número ni una cadena, lanzar un error
    raise ValueError(f"Unexpected value: {price}")

# Aplicar la función a la columna Precio
df_final['Precio'] = df_final['Precio'].apply(convert_price)
```

```python
columnas_a_convertir = ['Squad Number', 'Average Age', 'Ranking Points', 'Position']

for col in columnas_a_convertir:
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')
```

```python
def determinar_ganador(equipo):
    if equipo['home_score'] > equipo['away_score']:
        return equipo['home_team']
    elif equipo['home_score'] < equipo['away_score']:
        return equipo['away_team']
    else:
        return 'Draw'


df_resultados['winner'] = df_resultados.apply(determinar_ganador, axis=1)
df_resultados
```

```python
equipo1 = 'Brazil'
equipo2 = 'Spain'
torneo = 'todos'  # O el nombre específico del torneo, por ejemplo, 'World Cup'

# Si torneo es 'todos', incluimos todos los torneos en una lista
if torneo.lower() == 'todos':
    torneos = df_resultados['tournament'].unique()
else:
    torneos = [torneo]

# Filtrar para obtener solo los partidos entre Spain y Germany en los torneos especificados
partidos = df_resultados[(df_resultados['tournament'].isin(torneos)) &
                (((df_resultados['home_team'] == equipo1) & (df_resultados['away_team'] == equipo2)) |
                 ((df_resultados['home_team'] == equipo2) & (df_resultados['away_team'] == equipo1)))]

# Contar cuántos partidos ha ganado cada equipo
ganados_equipo1 = len(partidos[partidos['winner'] == equipo1])
ganados_equipo2 = len(partidos[partidos['winner'] == equipo2])
empates = len(partidos[partidos['winner'] == 'Draw'])

# Mostrar los resultados
if torneo.lower() == 'todos':
    print(f"Enfrentamientos entre {equipo1} y {equipo2} en todos los torneos:")
else:
    print(f"Enfrentamientos entre {equipo1} y {equipo2} en {torneo}:")
print(f"{equipo1} ha ganado {ganados_equipo1} partidos.")
print(f"{equipo2} ha ganado {ganados_equipo2} partidos.")
print(f"Ha habido {empates} empates.")
```

```python
def calculate_wins(df_resultados):
    home_wins = df_resultados[df_resultados['winner'] == df_resultados['home_team']]['home_team'].value_counts().rename('wins')
    away_wins = df_resultados[df_resultados['winner'] == df_resultados['away_team']]['away_team'].value_counts().rename('wins')
    return home_wins.add(away_wins, fill_value=0)
```

```python
def calculate_draws(df_resultados):
    home_draws = df_resultados[df_resultados['winner'] == 'Draw']['home_team'].value_counts().rename('draws')
    away_draws = df_resultados[df_resultados['winner'] == 'Draw']['away_team'].value_counts().rename('draws')
    return home_draws.add(away_draws, fill_value=0)
```

```python
def calculate_losses(df_resultados):
    home_losses = df_resultados[(df_resultados['winner'] != df_resultados['home_team']) & (df_resultados['winner'] != 'Draw')]['home_team'].value_counts().rename('losses')
    away_losses = df_resultados[(df_resultados['winner'] != df_resultados['away_team']) & (df_resultados['winner'] != 'Draw')]['away_team'].value_counts().rename('losses')
    return home_losses.add(away_losses, fill_value=0)
```

```python
def calculate_total_statistics(df_resultados):
    wins = calculate_wins(df_resultados)
    draws = calculate_draws(df_resultados)
    losses = calculate_losses(df_resultados)
    
    total_statistics = pd.concat([wins, draws, losses], axis=1).fillna(0).astype(int)
    return total_statistics

# Calculate total statistics by team
total_statistics = calculate_total_statistics(df_resultados)
total_statistics.head(50)
```
